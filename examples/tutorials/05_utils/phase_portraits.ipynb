```python
import expsolve as es
import expsolve.discretize.tensorgrid as ex
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
xr=np.array([[-10,10], [-10,10]])
n = [500,500]
[x, y] = ex.grid(n, xr)
z = x + 1j*y
```

```python
specsphaseportrait = lambda u: [(xr, es.utils.phase(u), mpl.colormaps['hsv'], torch.ones_like(x), 'none')]  # important not to use interpolation here
def drawphaseportrait(u):
    fig, ax = plt.subplots()
    ex.imshow(ax, imspecs=specsphaseportrait(u), bgcolor='black', xlim=xr[0], ylim=xr[1], vmin=0, vmax=1)
    plt.show()
```

```python
s = -2.5 + 5.0j
drawphaseportrait(1/(z-s))
```

```python
fz = torch.sin(z)-torch.sinh(z)
drawphaseportrait(fz)
```
